This notebook presents the interactive environment used in this project.

The environment is a grid-based puzzle where an agent must navigate through the grid, visiting targets in a specific order while respecting movement constraints such as walls and non-revisitable cells.

Below, you will find an interactive test that allows the user to manually play the game using the keyboard. This helps validate the environment dynamics and provides an intuitive understanding of the problem before applying learning algorithms.

```python
import time
import yaml
from src.environment import ZipEnvironment

# Read params from YAML file
with open("src/params.yaml", "r") as file:
    params = yaml.safe_load(file)
    
env = ZipEnvironment()
env.initializeGrid()

while True:
    # Agent has solved the puzzle
    if env.hasReachedEnd():
        time.sleep(0.5)
        break
    
    # Agent has reached a terminal state (end of episode)
    if env.hasReachedDeadEnd():
        time.sleep(0.5)
        env.reset()
    
    env.renderGame()
```

# Episode Test

In this section, we evaluate the environment through episodic interactions using a simple random policy.

The objective of this test is to:

- Validate the environment dynamics
- Ensure that the reward structure behaves as expected
- Observe termination conditions (goal reached or dead-end)
- Provide a baseline behavior before applying learning algorithms

At each step, the agent selects a random action from the available action space. Although this policy is not optimal, it is useful for debugging and verifying the correctness of the environment implementation.

```python
import time
import random
from src.environment import ZipEnvironment

num_episodes = 5
max_steps = 100

env = ZipEnvironment()
env.initializeGrid()

for ep in range(num_episodes):
    state = env.reset()
    total_reward = 0


    for step in range(max_steps):
        # Select a random action
        action = random.choice(list(env.ACTIONS.values()))

        # Take a step in the environment
        next_state, reward, done, _ = env.step(action)

        total_reward += reward
        state = next_state


        # Check if episode finished
        if done:
            print("Episode finished")
            break
        
        env.renderGame()
        time.sleep(0.0005)

    print(f"Total reward: {total_reward:.2f}")

env.closeRender()
```
